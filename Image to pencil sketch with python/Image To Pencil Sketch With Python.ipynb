```python
import cv2
```

```python
from PIL import Image as im
  
```

```python
import numpy as np
```

```python
img=cv2.imread("jerry.png")
img

```

```python
gray_filter=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
gray_filter
```

```python
gray_filter.shape
```

```python
gf = im.fromarray(gray_filter)
```

```python
gf
```

```python
invert=cv2.bitwise_not(gray_filter)
```

```python
invert
```

```python
invert.shape
```

```python
i=im.fromarray(invert)
```

```python
i
```

```python
blur=cv2.GaussianBlur(invert,(21,21),0)
```

```python
blur
```

```python
blur.shape
```

```python
b=im.fromarray(blur)
```

```python
b
```

```python
invertedblur=cv2.bitwise_not(blur)
```

```python
invertedblur
```

```python
invertedblur.shape
```

```python
ib=im.fromarray(invertedblur)
```

```python
ib
```

```python
sketch_filter=cv2.divide(gray_filter,invertedblur,scale=256.0)
```

```python
sketch_filter
```

```python
sketch_filter.shape
  
```

```python
sketch_filter=im.fromarray(sketch_filter)
```
